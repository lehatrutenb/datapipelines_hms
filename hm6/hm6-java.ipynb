```java
JavaRDD<Row> rdd = df.toJavaRDD();
```

```java
%%time

JavaPairRDD<Double, Tuple2<Double, Integer>> pairRDD = rdd
    .mapToPair(row -> {
        double protocolType = row.getDouble(1);  // Protocol Type at index 1
        double headerLength = row.getDouble(0);  // Header Length at index 0
        return new Tuple2<>(protocolType, new Tuple2<>(headerLength, 1));
    });

// Use aggregateByKey
JavaPairRDD<Double, Tuple2<Double, Integer>> aggregatedRDD = pairRDD
    .aggregateByKey(
        new Tuple2<>(0.0, 0),
        (acc, value) -> new Tuple2<>(acc._1 + value._1, acc._2 + value._2),
        (acc1, acc2) -> new Tuple2<>(acc1._1 + acc2._1, acc1._2 + acc2._2)
    );

// Calculate mean
JavaPairRDD<Double, Double> meanRDD = aggregatedRDD
    .mapToPair(tuple -> new Tuple2<>(
        tuple._1,  // Protocol Type
        tuple._2._1 / tuple._2._2  // Sum / Count
    ));

// Collect and print results
meanRDD.collect();
```
